```python
headers = {
        "X-RapidAPI-Key": "f961accce9mshf6c30b889f4c2acp1e66e7jsnf8cbb3e8e4e7",
        "X-RapidAPI-Host": "moviesdatabase.p.rapidapi.com"  
        }

# Creamos un diccionario (de momento para el 2010)
diccionario_comedy = {"tipo": [],
                    "titulo": [],
                    "año": [],
                    "mes": [],
                    "id":[]}

# La llamada contiene 4 keys (page, next, entries, results)
# Los datos que nos interesan estan en results por eso el bucle lo hacemos sobre llamada["results"] (que es una lista con los datos de las peliculas)
# Accedemos a cada una de las peliculas de la lista llamadas["results"]
for anho in range(2010,2025):
    for pag in range(1,3):
        url = f"https://moviesdatabase.p.rapidapi.com/titles?year={anho}&genre=Comedy&page={pag}"
        # Realiza una solicitud GET a la URL proporcionada y almacena la respuesta en 'llamada'
        consulta = requests.get(url, headers= headers) 
        # Convierte la consulta a formato json
        llamada = consulta.json()      
        for peliculas in llamada["results"]: 
            # Sacamos el valor de la clave tipo
            tipo = peliculas["titleType"]["text"]
            if tipo == "Movie" or tipo =="Short":
                # Lo añadimos al dic dentro de la clave tipo
                diccionario_comedy["tipo"].append(tipo)
                # Sacamos el valor de la clave id   
                id = peliculas["id"]     
                # Lo añadimos al dic dentro de la clave id
                diccionario_comedy["id"].append(id)
                # Sacamos el valor de la clave titulo
                titulo = peliculas["titleText"]["text"]
                # Lo añadimos al dic dentro de la clave titulo    
                diccionario_comedy["titulo"].append(titulo)
                # Como hemos detectado que algunas peliculas no tienen año tratamos esta excepcion
                try:
                    anho == peliculas["releaseDate"]["year"]
                    # Lo añadimos al dic dentro de la clave año   
                    diccionario_comedy["año"].append(anho)
                except:
                    # En caso de que no lo encuentre añadimos al dic dentro de la clave año el valor 
                    diccionario_comedy["año"].append(0000)
                # Como hemos detectado que algunas peliculas no tienen mes tratamos esta excepcion
                try:
                    # Sacamos el valor de la clave mes
                    mes = peliculas["releaseDate"]["month"]
                    # Lo añadimos al dic dentro de la clave mes   
                    diccionario_comedy["mes"].append(mes)
                except:
                    # En caso de que no lo encuentre añadimos al dic dentro de la clave mes el valor none
                    diccionario_comedy["mes"].append("none")
```

```python
import mysql.connector

# Conectarse a la base de datos
cnx = mysql.connector.connect(user='root', password='AlumnaAdalab',
                              host='127.0.0.1', database='proyecto2')

mycursor = cnx.cursor()

# SQL para insertar datos en la tabla peliculas
sql = "INSERT INTO oscars (id_Oscar, Año_oscar, Mejor_Pelicula, Mejor_director, Mejor_actor, Mejor_actriz) VALUES (%s, %s, %s, %s, %s, %s)"

# Datos a insertar
val1 = lista_sin_primero

val2 = [
    (7.7, 77, "Richard Donner", ["Chris Columbus", "Steven Spielberg"], "Los Goonies son un grupo de amigos que viven en Goon Docks, Astoria, pero sus casas han sido compradas y van a ser demolidas. Sin embargo, vivirán su última aventura en busca de un tesoro que pueda salvar el barrio.", "Aventura", "1h 54min", "Los Goonies"),
    (8.5, 85, "Director 2", ["Guionista 1", "Guionista 2"], "Sinopsis de Otra Pelicula", "Género de Otra Pelicula", "2h 10min", "Otra Pelicula"),
    # Agrega más detalles de películas correspondientes a val1
]

val = []

# Asegurarse de que las listas tengan la misma longitud
min_len = min(len(val1), len(val2))

# Procesar los datos para cada película
for i in range(min_len):
    tipo = val1[i][0]
    titulo = val1[i][1]
    fecha_estreno = val1[i][2]
    id_pelicula = val1[i][3]
    imdb = val2[i][0]
    tomatoes = val2[i][1]
    directores = ','.join(val2[i][2])
    guionistas = ','.join(val2[i][3])
    argumento = val2[i][4]
    genero = val2[i][5]
    duracion = val2[i][6]
    titulo2 = val2[i][7]
    
    val.append((tipo, titulo, fecha_estreno, id_pelicula, imdb, tomatoes, directores, guionistas, argumento, genero, duracion, titulo2))

try:
    mycursor.executemany(sql, val)
    cnx.commit()
    print(mycursor.rowcount, "registro/s insertado/s.")
except mysql.connector.Error as err:
    print("Error: {}".format(err))
finally:
    mycursor.close()
    cnx.close()
```
